```python
def plot_histogram(lld, counts):
    bin_width = lld[1] - lld[0]

    plt.rcParams.update({
        'font.size': 12,
        'axes.labelsize': 13,
        'axes.titlesize': 14,
    })

    plt.figure(figsize=(7.5, 4.8))
    plt.bar(
        lld,
        counts,
        width=bin_width ,   
        align='center',
        edgecolor='black',
        linewidth=0.9,
        color='teal',
    )
    
    plt.xlabel('Lower Level Discriminator (LLD) Voltage (V)')
    plt.ylabel('Counts')
    plt.title('SCA Response (Counts vs LLD Voltage)')
    plt.grid(alpha=0.6)
    plt.tight_layout()
    plt.show()

for lld, counts in zip([lld1, lld2, lld3, lld4, lld5], [rates1, rates2, rates3, rates4, rates5]):
    plot_histogram(lld, counts)
```
