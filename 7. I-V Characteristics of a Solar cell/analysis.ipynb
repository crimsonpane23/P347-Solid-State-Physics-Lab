### I-V Characteristics of Solar Cell


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Lamp source

```python
table1 = pd.read_csv(r'table1.csv')
table1['P0'] =  (table1['I0'].values)*(table1['V0'].values)
table1['P1'] =  (table1['I1'].values)*(table1['V1'].values)
table1['P2'] =  (table1['I2'].values)*(table1['V2'].values)
table1['P3'] =  (table1['I3'].values)*(table1['V3'].values)
table1['P4'] =  (table1['I4'].values)*(table1['V4'].values)
table1
```

```python
labels = {
    0: "No filter",
    1: "Filter 1 (pink)",
    2: "Filter 2 (green)",
    3: "Filter 3 (yellow)",
    4: "Filter 4 (blue)"
}

markers = ['o', 's', '^', 'D', 'v']
linestyles = ['-', '--', '-.', ':', (0, (3,1,1,1))]
color = ['black', 'purple', 'green', 'orange', 'blue']

plt.figure(figsize=(11,8))
plt.grid()

for i in range(5):
    V = table1[f'V{i}'].values
    I = table1[f'I{i}'].values
    P = table1[f'P{i}'].values
    
    mask = ~np.isnan(V) & ~np.isnan(I)
    V, I, P = V[mask], I[mask], P[mask]
    
    mpp_index = np.argmax(P)
    V_mp = V[mpp_index]
    I_mp = I[mpp_index]
    
    c = color[i]
 
    line, = plt.plot(V, I,
                     marker=markers[i],
                     linestyle=linestyles[i],
                     linewidth=2,
                     markersize=6,
                     label=labels[i],
                     color=c)
    
   
    plt.scatter(V_mp, I_mp, color=c, zorder=5)
    
    plt.plot([V_mp, V_mp], [0, I_mp],
             linestyle='--', color=c, alpha=0.6)
    
    plt.plot([0, V_mp], [I_mp, I_mp],
             linestyle='--', color=c, alpha=0.6)
 
    plt.annotate(f"MPP",
                 xy=(V_mp, I_mp),
                 xytext=(8, 8),
                 textcoords='offset points',
                 color=c,
                 fontsize=10)

plt.xlabel("Voltage (V)", fontsize=14)
plt.ylabel(r"Current $ (\mu A) $", fontsize=14)
plt.title("IV Characteristics of Solar Cell (Lamp source)", fontsize=16)

plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Max power output as a function of various filters for Sun and lamp as a source

```python
filters = ['No filter','Yellow', 'Pink', 'Blue', 'Green']
Pmax_lamp = [3870.75,2918.28,2379.65,1194.45,847.22]
Pmax_sun = np.array([280,199.9,139.58,87.22,122.75])

plt.figure(figsize=(8,6))
plt.plot(filters, Pmax_lamp, marker='o', linewidth=2)
plt.xlabel("Filter Type", fontsize=14)
plt.ylabel("Maximum Power ($\mu$W)", fontsize=14)
plt.title("Maximum Power for Different Optical Filters (Lamp as a source)")
plt.grid()
plt.show()
```
