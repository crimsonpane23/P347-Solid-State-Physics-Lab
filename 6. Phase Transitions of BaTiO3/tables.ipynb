# Study of Phase Transitions of BaTiO3


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
data = pd.read_csv('table1.csv')
C4 = data['C4'].values
R4 = data['R4'].values
f = data['f'].values
R3 = 1000           #ohm
C2 = 1000           #pF
r = (10.56e-3)/2    #m
t = 1.6e-3          #m
C1 = C2*R4/(R3/1000) #pF

data['C1'] = C1
R1 = R3*C4/C2       #ohm

data['R1'] = R1  #ohm

C0 = (8.854e-12)*(np.pi*(r**2)/t)/1e-12     #pF
print(C0,'pF')
data['Dielectric constant'] = C1/C0

gamma = 2*np.pi*f*C1*R1*1e-9
data['Dissipation factor'] = gamma
data
```

```python
#Dielectric constant vs Frequency
plt.figure(figsize=(10, 6))
plt.plot(f, data['Dielectric constant'], marker='o')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Dielectric Constant')
plt.title('Dielectric Constant vs Frequency')
plt.grid()
plt.show()
```

```python
#Dissipation factor vs Frequency
plt.figure(figsize=(10, 6))
plt.plot(f, data['Dissipation factor'], marker='o')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Dissipation Factor')
plt.title('Dissipation Factor vs Frequency')
plt.grid()
plt.show()
```

```python
#MLCC 
fc = np.array([7,10,15,20,25,30,35,40,45,50])
R4c = np.array([0.902,0.896,0.892,0.888,0.892,0.886,0.886,0.884,0.880,0.876]) 
R3C = 1.5
C1c = C2*R4c/R3C
print(C1c)

plt.figure(figsize=(10, 6))
plt.plot(fc, C1c, marker='o')
plt.xlabel('Frequency (kHz)')
plt.ylabel('C1 (pF)')
plt.title('C1 vs Frequency for MLCC')
plt.grid()
plt.show()
```

```python
#Disc capacitor
fd = np.array([3,5,7,10,15,20,25,30,35,40,45,50])
R4D = np.array([0.590,0.590,0.590,0.590,0.590,0.588,0.588,0.586,0.586,0.588,0.590,0.588])

C1D = C2*R4D/(R3/1000)
print(C1D)
plt.figure(figsize=(10, 6))
plt.plot(fd, C1D, marker='o')
plt.xlabel('Frequency (kHz)')
plt.ylabel('C1 (pF)')
plt.title('C1 vs Frequency for Disc Capacitor')
plt.grid()
plt.show()
```

```python
#Temperature variation for Different frquencies 
T = np.array([50,60,70,80,90,100,110,120,125,130,135,140,150,168])

#OP f = 15kHz
R4_15kHz = np.array([0.876,0.892,0.922,0.964,1.010,1.094,1.210,1.398,1.532,1.680,1.772,1.744,1.528,1.202])
C4_15kHz = np.array([350,350,400,500,500,600,600,650,750,950,600,1150,1150,400])
C1_15kHz = C2*R4_15kHz/(R3/1000)
E_15kHz = C1_15kHz/C0
print(np.round(E_15kHz,1))

#OP f = 25kHz
R4_25kHz = np.array([0.866,0.884,0.916,0.952,1.002,1.082,1.184,1.372,1.498,1.606,1.736,1.684,1.484,1.174])
C4_25kHz = np.array([200,250,250,250,350,350,450,400,600,600,850,1100,400])
C1_25kHz = C2*R4_25kHz/(R3/1000)
E_25kHz = C1_25kHz/C0
print(np.round(E_25kHz,1))


#OP f = 35kHz
R4_35kHz = np.array([0.862,0.878,0.910,0.948,0.996,1.066,1.170,1.348,1.460,1.592,1.720,1.682,1.466,1.170])
C4_35kHz = np.array([150,200,200,200,350,350,300,350,450,450,500,750,900,400])
C1_35kHz = C2*R4_35kHz/(R3/1000)
E_35kHz = C1_35kHz/C0
print(np.round(E_35kHz,1))

#OP f = 45kHz
R4_45kHz = np.array([0.858,0.874,0.906,0.942,0.996,1.060,1.166,1.334,1.440,1.556,1.690,1.624,1.428,1.152])
C4_45kHz = np.array([150,150,150,150,250,250,300,450,350,750,700,400])
C1_45kHz = C2*R4_45kHz/(R3/1000)
E_45kHz = C1_45kHz/C0
print(np.round(E_45kHz,1))
```

```python
#Plotting temperature variations of dielectric constant for various frequencies

plt.figure()
plt.plot(T, E_15kHz, label='15kHz', marker='x')
plt.plot(T, E_25kHz, label='25kHz', marker='o')
plt.plot(T, E_35kHz, label='35kHz', marker='v')
plt.plot(T, E_45kHz, label='45kHz', marker='*')
plt.xlabel('Temperature')
plt.ylabel('Dielectric constant')
plt.grid()
plt.legend()
plt.show()
```

```python
#Diffuseness parameter plot
E_max_15 = max(E_15kHz)

Tc = 135
print(T[-3:])
diffx = np.log(T[-3:] - Tc)
diffy_15 = np.log((E_15kHz[-3:]**(-1)) - (E_max_15**(-1)))
print(E_15kHz**(-1))

x = diffx
y = diffy_15
print(x)
n = len(x)
# --- Data Table ---
table_data = PrettyTable(["i", "x", "y", "x", "xy"])
for i in range(n):
    table_data.add_row([i+1, round(x[i], 3), round(y[i], 3), round(x[i]**2, 3), round(x[i]*y[i], 3)])
table_data.add_row(["", round(np.sum(x), 3), round(np.sum(y), 3), round(np.sum(x**2), 3), round(np.sum(x*y), 3)])
table_data.title = "Input Data"
print(table_data, "\n")

# --- Least Squares Calculation ---
Sx = np.sum(x)
Sy = np.sum(y)
Sxx = np.sum(x**2)
Sxy = np.sum(x*y)

delta = n*Sxx - Sx**2
a1 = (n*Sxy - Sx*Sy) / delta  # Slope
a0 = (Sxx*Sy - Sx*Sxy) / delta  # Intercept

# --- Error Analysis ---
y_fit = a0 + a1*x
residuals = y - y_fit
sigma_y = np.sqrt(np.sum(residuals**2) / (n-2))
sigma_a1 = sigma_y * np.sqrt(n / delta)
sigma_a0 = sigma_y * np.sqrt(Sxx / delta)

# --- Results Table ---
results = PrettyTable(["Parameter", "Value", "Uncertainty"])
results.title = "Fit Results & Statistical Errors"
results.add_row(["Slope (a)", f"{a1:.4f}", f"{sigma_a1:.4f}"])
results.add_row(["Intercept (a)", f"{a0:.4f}", f"{sigma_a0:.4f}"])
results.add_row(["_y (Std. Error)", f"{sigma_y:.4f}", ""])
results.add_row([" (delta)", f"{delta:.4f}", ""])
print(results, "\n")

# --- Fitted Values Table ---
fit_table = PrettyTable(["x", "y (measured)", "y (fit)"])
fit_table.title = "Measured vs Fitted Values"
for xi, yi, yfi in zip(x, y, y_fit):
    fit_table.add_row([round(xi, 3), round(yi, 3), round(yfi, 3)])
print(fit_table, "\n")

# --- Plotting ---
plt.figure(figsize=(7,5))
plt.scatter(x, y, color='navy', label='Measured Data', s=40)
plt.plot(x, y_fit, color='crimson', linewidth=2,
         label=f'Least Squares Fit:\n$y = ({a1:.5f} \pm {sigma_a1:.5f})x + ({a0:.4f} \pm {sigma_a0:.4f})$')
plt.xlabel(r"$ln(T - T_c)$", fontsize=12)
plt.ylabel(r"$ln(\frac{1}{\epsilon} - \frac{1}{\epsilon_{max}})$", fontsize=12)
plt.title("Diffuseness parameter (f = 15kHz)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
